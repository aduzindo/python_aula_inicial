```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def quadrado(x): # resulução de problema quadratica
    return x**2
```

```python
def cubo(x): # resulução de problema cúbica
    return x**3
```

```python
def bascara(a, b, c): # resulução de problema bascara
    x1=(-b+np.sqrt(b**2-4*a*c))/(2*a)
    x2=(-b-np.sqrt(b**2-4*a*c))/(2*a)
    return x1,x2
```

## Teste das Funções

```python
assert quadrado(4) == 16
assert quadrado(2) == 4
assert quadrado(3) ==9
assert quadrado(1) ==1
assert quadrado(5) == 25
```

```python
assert cubo(3) == 27
assert cubo(1) == 1
assert cubo(2) == 8
assert cubo(5) == 125
assert cubo(4) == 64
```

```python
assert bascara(1, 3, 2) == (-1, -2)
assert bascara(1, -2, 1) == (1, 1)
assert bascara(-2, -1, 3) == (-3/2, 1)
assert bascara(-2,-1,1)  ==  (-1, 1/2)
assert bascara(1,-3,-4) == (4, -1)
```

## Gráficos

```python
x=np.linspace(0,99) 
y=quadrado(x)
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,99)
y=cubo(x)
plt.plot(x,y)
plt.show()
```
